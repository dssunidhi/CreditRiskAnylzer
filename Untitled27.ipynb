#### Highlight the thresholds where customers become riskier
  - Calculate Debt-to-Income Ratio
  - Convert DAYS_BIRTH to age
  - Plot Default Rates Across Demographics

- Plots
  - Default rate by gender
  - Default rate by age group
  - Default rate by location

```python

client_detailed_data['DTI'] = (client_detailed_data['AMT_ANNUITY'] / client_detailed_data['AMT_INCOME_TOTAL']) * 100

# Convert DAYS_BIRTH to age
client_detailed_data['AGE'] = -client_detailed_data['DAYS_BIRTH'] // 365

# Plot Default Rates Across Demographics
plt.figure(figsize=(14, 8))

# Subplot 1: Default rate by gender
plt.subplot(2, 2, 1)
sns.barplot(x='CODE_GENDER', y='TARGET', data=client_detailed_data, estimator='mean', ci=None)
plt.title('Default Rate by Gender')

# Subplot 2: Default rate by age group
plt.subplot(2, 2, 2)
sns.histplot(x='AGE', hue='TARGET', data=client_detailed_data, multiple='stack', bins=20, kde=True)
plt.title('Default Rate by Age Group')

# Subplot 3: Default rate by location (assuming 'FLAG_OWN_REALTY' is a location indicator)
plt.subplot(2, 2, 3)
sns.barplot(x='FLAG_OWN_REALTY', y='TARGET', data=client_detailed_data, estimator='mean', ci=None)
plt.title('Default Rate by Location')

plt.tight_layout()
plt.show()
```

#### Heatmap Plot

```python
# Subset the data for relevant columns
columns_of_interest = ['TARGET', 'CODE_GENDER', 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY', 'CNT_CHILDREN', 'AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_ANNUITY', 'AMT_GOODS_PRICE']
subset_df = client_detailed_data[columns_of_interest]

# Calculate correlations between numeric variables
correlation_matrix = subset_df.corr()

# Plot heatmap for correlations
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Risk Heatmap - Numeric Variables')
plt.show()

# Plot heatmap for categorical variables
plt.figure(figsize=(10, 6))

# Convert categorical variables to numeric for correlation calculation
numeric_df = pd.get_dummies(subset_df.drop('TARGET', axis=1), drop_first=True)

# Concatenate numeric and target columns
corr_df = pd.concat([numeric_df, subset_df['TARGET']], axis=1)

# Calculate correlations
categorical_correlation_matrix = corr_df.corr()

sns.heatmap(categorical_correlation_matrix[['TARGET']], annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Risk Heatmap - Categorical Variables')
plt.show()

```

## Summary
-----------------

- The first heatmap shows correlations among numeric variables, highlighting potential relationships and patterns.

- The second heatmap focuses on the correlation between categorical variables (after one-hot encoding) and the target variable.

- Correlation values closer to 1 indicate a strong positive correlation, while values closer to -1 indicate a strong negative correlation.
